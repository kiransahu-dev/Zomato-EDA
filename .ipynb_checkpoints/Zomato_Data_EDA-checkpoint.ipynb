# 🍽️ Zomato Data Analysis (EDA)

## 📌 Project Overview
This project performs Exploratory Data Analysis (EDA) on the Zomato dataset to uncover insights about restaurant trends, pricing, ratings, and customer preferences.

## 🎯 Objectives
- Understand restaurant distribution and trends
- Analyze ratings and pricing patterns
- Identify popular cuisines and locations
- Extract actionable insights

## 🛠️ Tools Used
- Python
- Pandas
- Matplotlib
- Seaborn
- Jupyter Notebook


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Dataset load

```python
df = pd.read_csv('zomato.csv', encoding='latin-1')
df_country = pd.read_excel('Country-Code.xlsx')
```

```python
df.head()
```

```python
df_country.head()
```

```python
final_df = pd.merge(df, df_country, on='Country Code', how='left')
```

```python
final_df.head()
```

```python
final_df.shape
```

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_values = final_df.Country.value_counts().values
country_values
```

```python
# Pie Chart for Top-3 Country use Zomato
```

```python
plt.pie(country_values[:3],labels=country_names[:3], autopct='%1.2f%%')
plt.show()
```

## Observation: Zomato maximum transactions are from India next United States and next United Kingdom

```python
final_df.columns
```

```python
rating = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
rating
```

## Observation 2:
1. When rating is between 4.5 to 4.9 ---> Excelent
2. When rating is between 4.0 to 4.4 ---> Very Good
3. When rating is between 3.5 to 3.9 ---> Good
4. When rating is between 2.5 to 3.4 ---> Average
5. When rating is between 1.8 to 2.4 ---> Poor

```python
rating.head()
```

### Review details based on rating color

```python
plt.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating", y="Rating Count", hue="Rating color", data=rating, palette=['wheat', 'red', 'orange', 'yellow', 'green', 'green'])
```

### Observation 3:
1. Not Rated count is very high
2. Max number of rating between 2.5 to 3.4 i.e., "Average".


## Frequency Count of rating color

```python
## Count Plot
sns.countplot(data=rating, x='Rating color',hue='Rating color', palette=['wheat', 'red', 'orange', 'yellow', 'green', 'green'],legend=False)
```

### Create a pie chart of top 5 city distributions

```python
final_df.columns
```

```python
City_count = final_df['City'].value_counts().values
City_labels = final_df['City'].value_counts().index
City_count
City_labels
```

```python
plt.pie(City_count[:5], labels=City_labels[:5], autopct="%1.2f%%")
plt.show()
```

## Observation:

### Find top 10 cuisines ordered

```python
final_df.columns
```

```python
final_df.Cuisines.value_counts().values
top_10_cuisines = final_df.Cuisines.value_counts().values
top_10_cuisines
top_10_cuisines_labels = final_df.Cuisines.value_counts().index
top_10_cuisines_labels
```

```python
plt.pie(top_10_cuisines[:10], labels=top_10_cuisines_labels[:10],autopct='%1.2f%%')
plt.show()
```

## Observation:
1. Most Ordered Cuisines is North Indian
2. Then North Indian, Chinese
3. Chinese and Fast Food are similar number of order
